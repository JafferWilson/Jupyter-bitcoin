```python
%matplotlib inline
import pandas as pd
from fbprophet import Prophet

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('BTCE-USDBTC.csv')

df.head(5)
```

```python
df['Date'] = pd.DatetimeIndex(df['Date'])
df.dtypes
```

```python
df2 = df.rename(columns={'Date':'ds', 'Volume (Currency)': 'y'})
df2.head()
```

```python
df2 = df2[['ds','y']]
```

```python
ax = df2.set_index('ds').plot(figsize=(12, 8))
ax.set_ylabel('Daily volume in currency usd to btc')
ax.set_xlabel('Date')

plt.show()
```

```python
my_model = Prophet(interval_width=0.95)
my_model.fit(df2)
```

```python
future_dates = my_model.make_future_dataframe(periods=365, freq='D')
future_dates.tail()
```

```python
forecast = my_model.predict(future_dates)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
my_model.plot(forecast,
              uncertainty=True)
```

```python
my_model.plot_components(forecast)
```
